```python

 df_1 = pd.DataFrame(df_sorted_name_desc)
df_1.set_index('Dates', inplace=True)
```

```python

print(df_1.dtypes)
```

```python

from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error

# Define the SARIMA model parameters
order = (1, 1, 1)  # ARIMA(p, d, q) parameters
seasonal_order = (1, 1, 1, 12)  # SARIMA(P, D, Q, s) parameters

# Split data into train and test sets
train_size = int(len(df_1) * 0.8)
train, test = df_1[:train_size], df_1[train_size:]
```

```python

# Fit SARIMA model to the training data
model = SARIMAX(train, order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
result = model.fit()

# Forecast future values
forecast = result.get_forecast(steps=len(test))
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Plot actual vs. forecasted values
plt.figure(figsize=(14, 6))
plt.plot(train.index, train, label='Training Data', color='c')
plt.plot(test.index, test, label='Test Data', color='g')
plt.plot(test.index, forecast_mean, label='Forecast')
plt.fill_between(test.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='k', alpha=0.1)
plt.title('SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Evaluate the model
mse = mean_squared_error(test, forecast_mean)
print('Mean Squared Error:', mse)
```
